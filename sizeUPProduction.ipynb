```python
aLevel = pd.read_csv("/Users/sadiqkhawaja/Desktop/Brackets/WhereamI/sizeUps/AQA-A-LEVEL-STATS-JUN-2023.csv")
n = 8
aLevel.drop(index = aLevel.index[:n],inplace=True)
aLevel.drop(aLevel.tail(8).index, inplace=True)
aLevel.drop(index = aLevel.index[:2], inplace = True)
aLevel.drop(aLevel.columns[0], axis = 1, inplace = True)
aLevel.drop(aLevel.columns[9:22], axis = 1, inplace = True)
aLevel
```

```python
subjects = aLevel['Unnamed: 1']
subjects = subjects.tolist()
subjects
```

```python
results = aLevel.loc[:,'Unnamed: 3':'Unnamed: 9']
results = results.values.tolist()
results
```

```python
resultsAbsoluteValueArr = []
for j in range (0,len(results)):
    resultsAbsoluteValue = []
    for i in range(0,len(results[j])):
        if results[j][i] == '0~':
            results[j][i] = 0
        results[j][i] = float(results[j][i])
        if i != 0:
            resultsAbsoluteValue.append(float(format(results[j][i]-results[j][i-1],".2f")))
        else:
            resultsAbsoluteValue.append(float(format(results[j][i],".2f")))
    resultsAbsoluteValueArr.append(resultsAbsoluteValue)
resultsAbsoluteValueArr  
```

```python
def getSubjectandResults(subject,subjects,resultsAbsoluteValueArr):
    for i in range (0,len(subjects)):
        if subject in subjects[i]:
            return subjects[i],resultsAbsoluteValueArr[i]

def getGrade(subjectGrade, specificResults):
    grades = ["A*","A","B","C","D","E","U"]
    for i in range (0,len(grades)):
        if grades[i] == subjectGrade:
            return  i
    else:
        return "Not Valid","Not valid"

def drawChart(subject,specificResults,gradeIndex):
    grades = ["A*","A","B","C","D","E","U"]
    bars = plt.bar(grades, specificResults)
    bars[gradeIndex].set_color('red')
    plt.title(subject)
    plt.show()

def getChartSuperFunction(subject,subjectGrade,subjects,resultsAbsoluteValueArr):
    subject, specificResults = getSubjectandResults(subject,subjects,resultsAbsoluteValueArr)
    gradeIndex  = getGrade(subjectGrade, specificResults)
    drawChart(subject, specificResults, gradeIndex)

getChartSuperFunction("COMPUTER SCIENCE","E",subjects,resultsAbsoluteValueArr)
```
